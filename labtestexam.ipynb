```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

aqi_readings = np.random.randint(50,120, size=(1440))
noice = np.random.normal(0,10,1440)

aqi_readings_noice_data= aqi_readings + noice

print(aqi_readings_noice_data)

b,a=signal.butter(3,0.06)
smooth_aqi_data=signal.filtfilt(b,a,aqi_readings_noice_data)

avg=np.mean(smooth_aqi_data.reshape(24,60), axis=1)

print(avg)

x1=np.arange(1440)
y1=aqi_readings_noice_data
plt.subplot(1,3,1)
plt.plot(x1,y1,label="aqi readings with noice", color="green")
plt.xlabel("time")
plt.ylabel("aqi readings with noice")

x2=np.arange(1440)
y2=smooth_aqi_data
plt.subplot(1,3,2)
plt.plot(x2,y2,label="smooth aqi readings", color="blue")
plt.xlabel("time")
plt.ylabel("smooth aqi readings")

x3=np.arange(0,1440,60)
y3=avg
plt.subplot(1,3,3)
plt.plot(x3,y3,label="avg of aqi readings", color="black", marker='.')

plt.xlabel("time")
plt.ylabel("avg aqi readings")

plt.suptitle("Aqi index data")

plt.legend()
plt.show()

def fun():
    s=-1
    cnt=0
    
    for i in range(0,1440):
        if smooth_aqi_data[i]>= 100:
            cnt+=1
        else:
            cnt=0
        if cnt==15:
            s=i-15
            break
            
        
        
            









```
